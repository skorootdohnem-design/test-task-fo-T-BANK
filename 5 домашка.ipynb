```python
quily = germany_top.InvoiceNo.quantile(0.8)
germany_top1 = germany_top.query(f'InvoiceNo  > {quily}').CustomerID
retail
```

```python
top_retail_germany = retail[retail['CustomerID'].isin(germany_top1)]
```
