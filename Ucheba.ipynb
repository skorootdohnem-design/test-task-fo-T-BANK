```python
import pandas as pd
from datetime import datetime
```

```python
df = pd.read_csv('D:\\Программы\\Karpov.Courses - Аналитик данных (2022)\\Часть 1\\Задания\\Ноутбуки и датасеты\\Ноутбуки и датасеты\\lesson_1_data.csv' , sep = ';')
```

```python
df = df.rename(columns = {'Номер': 'number',
                'Дата создания': 'create_date',
                'Дата оплаты': 'payment_day' ,
                'Title': 'title' ,
                'Статус': 'status' ,
                'Заработано': 'money' ,
                'Город': 'city' ,
                'Платежная система': 'payment_system'})
```

```python
all_money = df.money.sum()
```

```python
money_by_city = df \
    .groupby(['title', 'city'], as_index = False) \
    .aggregate({'money': 'sum'}) \
    .sort_values('money', ascending = False)
```

```python
money_by_city.money.sum()
```

```python
money_by_city.to_csv('money_by_city.to_csv', index=False)
```

```python
money_gr = df \
    .query("status == 'Завершен'") \
    .groupby(['title', 'city'], as_index = False) \
    .aggregate({'money': 'sum', 'number' : 'count'}) \
    .sort_values('money', ascending = False) \
    .rename(columns={'number': 'succes_order'})
```

```python
today_day = datetime.today().strftime('%Y-%m-%d')
file_name = 'money_title_{}.csv'
file_name.format(today_day)
```
