```python
df = df.rename(columns={'Номер': 'number',
               'Дата создания': 'create_date',
               'Дата оплаты': 'payment_date',
               'Title': 'title',
               'Статус': 'status',
               'Заработано': 'money',
               'Город': 'city',
               'Платежная система': 'payment system'})
```

```python
df
```

```python
all_money = df.money.sum()
all_money
```

```python
df.title.unique()
```

```python
money_buy_city = df \
    .groupby(['title', 'city'], as_index=False)\
    .aggregate({'money': 'sum'})\
    .sort_values('money', ascending=False)
```

```python
money_buy_city
```

```python
money_buy_city.to_csv('D:\\Программы\\Karpov.Courses - Аналитик данных (2022)\\Часть 1\\Задания\\Ноутбуки и датасеты\\Ноутбуки и датасеты\\money_buy_city.csv', index=False)
```

```python
money_buy_city.to_csv('money_buy_city.csv', index=False, encoding='cp1251')
```

```python
money_title = df\
    .query("status == 'Завершен'") \
    .groupby(['title'], as_index=False)\
    .aggregate({'money': 'sum', 'number' : 'count'})\
    .sort_values('money', ascending=False)\
    .rename(columns={'number': 'succes_orders'})
```

```python
money_title
```

```python
today_day = datetime.today().strftime('%Y-%m-%d')
file_name = 'money_title_{}.csv'
file_name = file_name.format(today_day)
```

```python
if int(money_title.money.sum()) == int(all_money):
    print('ok File {} is written.'.format(file_name))
    money_title.to_csv(file_name, index=False)
else:
    print('no_ok')
```
