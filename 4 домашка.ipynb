```python
number = [1, -2, 3, 0, -3, 2]
positive_number = []

for num in number:
    if num > 0:
        positive_number.append(num)
        continue
    elif num == 0:
        break
```

```python
numbers2 = [2, 5, 8, 1, 10, 0, 12, 0, 14, 12]
even_numbers = []
for num in numbers2:
    if num // 2:
        even_numbers.append(num)
        continue

    elif num == 0:
        continue

    elif num // 2:
            even_numbers.append(num)
            break

print(even_numbers)
```
