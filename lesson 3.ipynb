```python
user_df = df[['tc', 'art_sp']]
```

```python
user_df = user_df.rename(columns={'tc': 'user_id', 'art_sp': 'brand_info'})
```

```python
user_df.head()
```

```python
brand_name = 'MARAVILLA 500 G Store_Brand'
brand_name.split(' ')[-1]
```

```python
user_df['brand_name'] = user_df['brand_info'].apply(lambda x: x.split(' ')[-1])
```

```python
user_df.head()
```

```python
df.head()
```

```python
user_purchases = user_df.groupby('user_id', as_index=False)\ # люди которые купили больше 5
    .agg({'brand_name': 'count'})\
    .rename(columns={'brand_name': 'purchases'})\
    .query('purchases >= 5')
# люди которые купили больше 5
```

```python
user_purchases.head()
```

```python
users_unique_brand = user_df.groupby('user_id', as_index=False)\
    .agg({'brand_name': pd.Series.nunique})\
    .rename(columns={'brand_name': 'unique_brands'})
#люди с уникальным брендом
users_unique_brand.head()
```

```python
lovely_brand_purchases_df = user_df.groupby(['user_id', 'brand_name'], as_index=False)\
    .agg({'brand_info': 'count'})\
    .sort_values(['user_id', 'brand_info'], ascending=[False, False])\
    .groupby('user_id')\
    .head(1)\
    .rename(columns={'brand_name': 'lovely_brand', 'brand_info': 'lovely_brand_purchases'})
# покупки для каждого пользовотеля любимого бренда
```

```python
loyality_df = user_purchases\
    .merge(users_unique_brand, on='user_id')\
    .merge(lovely_brand_purchases_df, on='user_id')
```

```python
loyality_df.head()
```

```python
Loyal_users = loyality_df[loyality_df.unique_brands == 1]
```

```python
loyality_df['loyalty_score'] = loyality_df.lovely_brand_purchases/ loyality_df.purchases
```
